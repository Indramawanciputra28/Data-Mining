```python
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score

# Fungsi untuk menampilkan proses preprocessing
def show_preprocessing_steps(img_path):
    # Baca gambar asli
    original_img = cv2.imread(img_path)
    if original_img is None:
        print(f"Gagal membaca gambar: {img_path}")
        return

    # 1. Crop ke rasio 3:2
    h, w = original_img.shape[:2]
    target_ratio = 3/2  # Diubah menjadi 3:2 sesuai permintaan
    current_ratio = w/h

    if current_ratio > target_ratio:
        # Crop lebar
        new_w = int(h * target_ratio)
        start_x = (w - new_w) // 2
        cropped_img = original_img[:, start_x:start_x+new_w]
    else:
        # Crop tinggi
        new_h = int(w / target_ratio)
        start_y = (h - new_h) // 2
        cropped_img = original_img[start_y:start_y+new_h, :]

    # 2. Resize ke 1000x400 piksel
    resized_img = cv2.resize(cropped_img, (1000, 400))

    # 3. Konversi ke grayscale
    gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)

    # 4. Gaussian Blur
    blurred_img = cv2.GaussianBlur(gray_img, (5, 5), 0)

    # 5. CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahe_img = clahe.apply(blurred_img)

    # Tampilkan hasil setiap tahap
    plt.figure(figsize=(18, 6))  # Diperbesar untuk menampilkan lebih banyak gambar

    plt.subplot(1, 6, 1)
    plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
    plt.title(f'Original\n{w}x{h}')
    plt.axis('off')

    plt.subplot(1, 6, 2)
    plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    plt.title(f'Cropped 3:2\n{cropped_img.shape[1]}x{cropped_img.shape[0]}')
    plt.axis('off')

    plt.subplot(1, 6, 3)
    plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
    plt.title('Resized\n1000x400')
    plt.axis('off')

    plt.subplot(1, 6, 4)
    plt.imshow(gray_img, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.subplot(1, 6, 5)
    plt.imshow(blurred_img, cmap='gray')
    plt.title('Gaussian Blur')
    plt.axis('off')

    plt.subplot(1, 6, 6)
    plt.imshow(clahe_img, cmap='gray')
    plt.title('CLAHE')
    plt.axis('off')

    plt.suptitle('Proses Preprocessing (Crop 3:2 + Resize 1000x400 + Grayscale + Gaussian Blur + CLAHE)', y=1.05)
    plt.tight_layout()
    plt.show()

    return clahe_img  # Mengembalikan gambar yang sudah diproses untuk ekstraksi fitur

# Fungsi untuk load dataset dan ekstraksi fitur GLCM
def load_dataset(base_path):
    categories = ['Hawar', 'Karat', 'Sehat']
    features = []
    labels = []
    filenames = []

    for idx, category in enumerate(categories):
        path = os.path.join(base_path, category)
        # Ambil satu gambar untuk contoh preprocessing
        example_img = os.listdir(path)[0]
        show_preprocessing_steps(os.path.join(path, example_img))

        for img_name in os.listdir(path):
            img_path = os.path.join(path, img_name)

            img = cv2.imread(img_path)
            if img is None:
                print(f"Gagal membaca gambar: {img_path}")
                continue

            # 1. Crop ke rasio 3:2
            h, w = img.shape[:2]
            target_ratio = 3/2
            current_ratio = w/h

            if current_ratio > target_ratio:
                new_w = int(h * target_ratio)
                start_x = (w - new_w) // 2
                cropped = img[:, start_x:start_x+new_w]
            else:
                new_h = int(w / target_ratio)
                start_y = (h - new_h) // 2
                cropped = img[start_y:start_y+new_h, :]

            # 2. Resize ke 1000x400
            resized = cv2.resize(cropped, (1000, 400))

            # 3. Konversi ke grayscale
            gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

            # 4. Gaussian Blur
            blurred = cv2.GaussianBlur(gray, (5, 5), 0)

            # 5. CLAHE
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
            processed_img = clahe.apply(blurred)

            # Ekstraksi fitur GLCM dari gambar yang sudah diproses (setelah CLAHE)
            glcm = graycomatrix(processed_img, distances=[5], angles=[0], levels=256, symmetric=True, normed=True)
            contrast = graycoprops(glcm, 'contrast')[0, 0]
            homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
            energy = graycoprops(glcm, 'energy')[0, 0]
            correlation = graycoprops(glcm, 'correlation')[0, 0]
            entropy = -np.sum(glcm[:,:,0,0] * np.log(glcm[:,:,0,0] + (glcm[:,:,0,0]==0)))

            # Simpan data
            filenames.append(img_name)
            features.append([contrast, homogeneity, energy, correlation, entropy])
            labels.append(idx)  # 0: Hawar, 1: Karat, 2: Sehat

    return np.array(features), np.array(labels), filenames, categories

# Path dataset
base_path = '/content/drive/MyDrive/Dataset Jagungg/Jagung'

# Load dataset dan ekstraksi fitur
print("Memuat dataset dan mengekstraksi fitur GLCM...")
features, labels, filenames, class_names = load_dataset(base_path)

# Buat DataFrame untuk fitur GLCM
feature_names = ['Contrast', 'Homogeneity', 'Energy', 'Correlation', 'Entropy']
df_features = pd.DataFrame(features, columns=feature_names)
df_features.insert(0, 'Nama File', filenames)
df_features['Label'] = labels
df_features['Label Name'] = df_features['Label'].apply(lambda x: class_names[x])

# Tampilkan 10 sampel acak dari fitur yang telah diekstraksi
print("\nTabel Fitur GLCM:")
pd.set_option('display.float_format', '{:.4f}'.format)
display(df_features.sample(10, random_state=42))

# Split dataset dengan test_size 0.2 dan random_state 42 untuk hasil yang lebih konsisten
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=42)

# Fungsi untuk evaluasi model
def evaluate_model(model, X_test, y_test, class_names, title):
    y_pred = model.predict(X_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title(f'Confusion Matrix {title}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Classification Report
    report = classification_report(y_test, y_pred, target_names=class_names, output_dict=True)
    accuracy = report['accuracy']
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1 = report['weighted avg']['f1-score']

    return accuracy, precision, recall, f1

# ==============================================
# EVALUASI KNN dengan parameter yang dioptimalkan
# ==============================================
k_values = [3, 5, 7, 9, 11]  # Mengurangi range nilai k untuk fokus pada nilai optimal
knn_results = []

print("\n=== EVALUASI KNN ===")
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance', metric='manhattan')  # Menambahkan parameter tambahan
    knn.fit(X_train, y_train)

    # Evaluasi
    acc, prec, rec, f1 = evaluate_model(knn, X_test, y_test, class_names, f'KNN (k={k})')

    knn_results.append({
        'k': k,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1
    })

# Tampilkan hasil KNN
knn_results_df = pd.DataFrame(knn_results)
print("\nHASIL EVALUASI KNN:")
display(knn_results_df)

# ==============================================
# EVALUASI LOGISTIC REGRESSION dengan parameter yang dioptimalkan
# ==============================================
print("\n=== EVALUASI LOGISTIC REGRESSION ===")
logreg = LogisticRegression(max_iter=1000, random_state=42, C=1.0, solver='lbfgs', multi_class='auto')
logreg.fit(X_train, y_train)

# Evaluasi
logreg_acc, logreg_prec, logreg_rec, logreg_f1 = evaluate_model(
    logreg, X_test, y_test, class_names, 'Logistic Regression')

# Tampilkan hasil Logistic Regression
logreg_results = pd.DataFrame({
    'Model': ['Logistic Regression'],
    'Accuracy': [logreg_acc],
    'Precision': [logreg_prec],
    'Recall': [logreg_rec],
    'F1-Score': [logreg_f1]
})
print("\nHASIL EVALUASI LOGISTIC REGRESSION:")
display(logreg_results)

# ==============================================
# PERBANDINGAN MODEL
# ==============================================
print("\n=== PERBANDINGAN MODEL ===")

# Ambil akurasi terbaik KNN
best_knn = knn_results_df.loc[knn_results_df['Accuracy'].idxmax()]
comparison = pd.DataFrame({
    'Model': ['KNN (k={})'.format(best_knn['k']), 'Logistic Regression'],
    'Accuracy': [best_knn['Accuracy'], logreg_acc],
    'Precision': [best_knn['Precision'], logreg_prec],
    'Recall': [best_knn['Recall'], logreg_rec],
    'F1-Score': [best_knn['F1-Score'], logreg_f1]
})

print("\nPerbandingan Model Terbaik:")
display(comparison)

# Visualisasi perbandingan
plt.figure(figsize=(10, 6))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, best_knn[metrics].values, width, label='KNN')
rects2 = ax.bar(x + width/2, [logreg_acc, logreg_prec, logreg_rec, logreg_f1], width, label='Logistic Regression')

ax.set_ylabel('Score')
ax.set_title('Perbandingan KNN dan Logistic Regression')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

fig.tight_layout()
plt.show()
```
