```python
# [1] Mempersiapkan library
import numpy as np  # Import library Numpy untuk operasi array dan komputasi numerik
import pandas as pd  # Import library Pandas untuk manipulasi dan analisis data berbasis tabel
import matplotlib.pyplot as plt  # Import library Matplotlib untuk membuat visualisasi data
import sklearn  # Import library Scikit-learn untuk algoritma machine learning dan preprocessing data
# [2] Memanggil Dataset
dataset = pd.read_excel('Data_Penjualan.xlsx')  # Membaca file Excel yang berisi dataset
x = dataset.iloc[:, :-1].values  # Memilih semua baris dan semua kolom kecuali kolom terakhir (fitur independen)
y = dataset.iloc[:, -1].values  # Memilih semua baris dan hanya kolom terakhir (fitur dependen/target)

# [5] Menampilkan kolom dataset
dataset.keys()  # Menampilkan nama kolom yang ada pada dataset

# [6] Menampilkan dimensi dataset
dataset.shape  # Mengembalikan jumlah baris dan kolom dalam dataset (baris, kolom)

# [7] Menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)  # Membuat DataFrame untuk menampilkan data
dataku.head()  # Menampilkan 5 baris pertama dari dataset untuk mendapatkan gambaran data

# [8] Split Dataset menjadi Training Set dan Testing Set
from sklearn.model_selection import train_test_split  # Mengimpor fungsi train_test_split dari sklearn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
# Membagi dataset menjadi training set dan testing set:
# - x_train, y_train: digunakan untuk melatih model
# - x_test, y_test: digunakan untuk menguji performa model
# - test_size=0.2: 20% data digunakan untuk testing, 80% untuk training
# - random_state=0: agar hasil pembagian dataset selalu konsisten
```

```python
# [9] Melakukan Fitting Simple Linear Regression pada Training Set
from sklearn.linear_model import LinearRegression  # Mengimpor kelas LinearRegression dari sklearn
regressor = LinearRegression()  # Membuat objek regresor linear
regressor.fit(x_train, y_train)  # Melatih model pada data training (x_train, y_train)

# [10] Memprediksi hasil Test-Set
y_pred = regressor.predict(x_test)  # Menggunakan model yang telah dilatih untuk memprediksi hasil data uji (x_test)

# [12] Visualisasi Data
plt.scatter(dataku.BiayaProduksi, dataku.NilaiPenjualan)  # Membuat scatter plot dari Biaya Produksi dan Nilai Penjualan
plt.xlabel("Biaya Produksi")  # Menambahkan label pada sumbu-x
plt.ylabel("Nilai Penjualan")  # Menambahkan label pada sumbu-y
plt.title("Grafik Biaya Produksi vs Nilai Penjualan")  # Memberikan judul pada grafik
plt.show()  # Menampilkan grafik
```

```python
# [13] Visualisasi Hasil Prediksi Pada Training-Set

# Ukuran Plot
plt.figure(figsize=(10, 8))  # Mengatur ukuran figure plot menjadi 10x8 inci

# Biru adalah data observasi
plt.scatter(x_train, y_train, color='blue')  # Membuat scatter plot data asli dari training set dengan warna biru

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color='red')
# Membuat garis prediksi regresi linier menggunakan data training dengan warna merah

# Memberi Judul dan Label
plt.title('Biaya Produksi terhadap Penjualan (Training Set)')  # Judul grafik
plt.xlabel('Biaya Produksi')  # Label sumbu-x
plt.ylabel('Penjualan')  # Label sumbu-y

# Menampilkan plot
plt.show()  # Menampilkan grafik

```

```python
# [14] Visualisasi Hasil Prediksi Pada Test-Set

# Ukuran Plot
plt.figure(figsize=(10, 8))  # Mengatur ukuran plot menjadi 10x8 inci agar lebih besar dan mudah dibaca

# Biru adalah data observasi
plt.scatter(x_test, y_test, color='blue')  # Membuat scatter plot data asli dari test set dengan warna biru

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_test, regressor.predict(x_test), color='red')
# Membuat garis regresi linier berdasarkan prediksi hasil model dengan warna merah

# Memberi Judul dan Label
plt.title('Biaya Produksi terhadap Penjualan (Test Set)')  # Judul grafik untuk test set
plt.xlabel('Biaya Produksi')  # Label sumbu-x
plt.ylabel('Penjualan')  # Label sumbu-y

# Menampilkan plot
plt.show()  # Menampilkan grafik
```
